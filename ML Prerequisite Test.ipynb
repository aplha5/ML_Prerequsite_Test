# 1st question

Given two matrices of dimensions (n,k) i.e, n rows and k columns and (k,m) i.e, k rows and m columns. Here after doing the matrix multiplication the dimensions of the resulting matrix will be (n,m). We find the elements in the resulting matrix by doing the dot product of the correspoding row of the first matrix and column of the second matrix which means that for every element in the resulting matrix we are doing k multiplications(number of colums of 1st matrix or number of rows of the second matrix ) and to add them k-1 additions(to add n numbers we use addition n-1 times). The total number of elements in the resulting matrix are n * m so therefore the total number of multiplications used for matrix multplication are n * m * k and the total number of additions are n * m * (k-1)

```python
import numpy as np
import timeit
n = int (input())
k=int (input())
m=int(input())
a=[[0 for _ in range(k)]for _ in range(n)]
b=[[0 for _ in range(m)]for _ in range(k)]
for i in range(n):
    for j in range(k):
        a[i][j]=int(input())
for z in range(k):
    for p in range(m):
        b[z][p]=int(input())
def listoflist(a,b):
    c=[[0 for _ in range(len(b[0]))]for _ in range(len(a))]
    for i in range(len(a)):
        for j in range(len(b[0])):
            for k in range(len(b)):
                c[i][j]+=a[i][k]*b[k][j]
    return c;
def use_numPy(a,b):
    return np.dot(a,b);
s_l=timeit.default_timer()
c=listoflist(a,b)
e_l=timeit.default_timer()
s_n=timeit.default_timer()
d=use_numPy(a,b)
e_n=timeit.default_timer()
print (c)
print(d)
if e_l-s_l>e_n-s_n:
    print ("listoflist multipclication is faster");
elif e_l-s_l<e_n-s_n :
    print ("numpy if faster")
else :
    print ("both are equal")
    
```

The numpy multiplication is faster than the list of list multiplications approach becasue numpy uses the optimised algorithms and libraries and also it vectorization which avoids the overhead of loops and also it used fixed size data types so numpy multiplication is faster

# 3rd Question

```python
import numpy as np
def native_median(arr):
    n=len(arr)
    for i in range(n//2+1):
        maxi=i
        for j in range(i+1,n):
            if(arr[j]>arr[maxi]):
                maxi=j
        arr[i],arr[maxi]=arr[maxi],arr[i]
    if n%2==0:
        return (arr[n//2]+arr[n//2-1])/2
    else:
        return arr[n//2]
def meadian_usingsorting(arr):
    arr.sort()
    n=len(arr)
    if(n%2==0):
        return (arr[n//2]+arr[n//2-1])/2
    else :
        return arr[n//2]
arr = input().split()
arr=np.array(arr).astype(float)
median1= native_median(arr)
median2=meadian_usingsorting(arr)
print("Native Median:", median1)
print("Median using sorting:", median2)

```

In the above native meadian method we used the method of calculating the n/2th highest elelment and it takes a time complexity of (n/2)(n) that is of O(n^2). But in the second case we used sorting technique and the time complexity of sorting is O(nlogn) so the time complexity of meadian_usingsorting methof is O(nlogn)

# 4th Question

Given equation is  f(x, y) = x^2*y + y^3 sin(x) \.

Partial derivatives of the equation are:

\begin{align}
    \frac{\partial f}{\partial x} &= 2xy + y^3cos(x) \\
    \frac{\partial f}{\partial y} &= x^2 + 3y^2 \sin(x)
\end{align}

The gradient is written as: 
\begin{equation}
    \nabla f(x, y) = \begin{bmatrix}
        \frac{\partial f}{\partial x} \\
        \frac{\partial f}{\partial y}
    \end{bmatrix}
\end{equation}

Gradient:

\begin{equation}
    \nabla f(x, y) = \begin{bmatrix}
        2xy + y^3 \cos(x) \\
       x^2+ 3y^2 \sin(x)
    \end{bmatrix}
\end{equation}

Differentiation of x^n is n*(x^(n-1)) and differentiation of sin(x) is cos(x)

# 8th Question

```python
class Student:
    def __init__(self,rollnumber,name):
        self.roll_number=rollnumber
        self.name=name;
        self.marks={}
    def add_marks(self,subject,marks):
        self.marks[subject]=marks
class Branch:
    def __init__(self,name):
        self.name=name
        self.students=[]
    def add_students(self,student):
        self.students.append(student)
class Year:
    def __init__(self,year):
        self.year=year
        self.branches=[]
    def add_branches(self,branch):
        self.branches.append(branch)
record_of_students=[]
while True:
    year=int(input("Enter year (or 0 to end):"))
    if year==0:
        break
    year_obj=Year(year)
    while True:
        branch=input("Enter branch (or done to end):")
        if branch=='done':
            break
        branch_obj=Branch(branch)
        while True:
            rollnumber=int(input("Enter roll number (or 0 to end):"))
            if rollnumber==0:
                break
            name=input("Enter name:")
            student_obj=Student(rollnumber,name)
            while True:
                subject=input("Enter subject(or done to end):")
                if subject=='done':
                    break
                marks=int(input("Enter marks :"))
                student_obj.add_marks(subject,marks)
        branch_obj.add_students(student_obj)
    year_obj.add_branches(branch_obj)
    record_of_students.append(year_obj)
for year_obj in record_of_students:
    print(year_obj.year)
    for branch_obj in year_obj.branches:
        print(f"  - Branch: {branch_obj.name}")
        for student_obj in branch_obj.students:
            print(f"    - Roll Number: {student_obj.roll_number}, Name: {student_obj.name}")
            print(f"      - Marks: {student_obj.marks}")            
        
```
